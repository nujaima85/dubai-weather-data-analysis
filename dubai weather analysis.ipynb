# data predicton 


```python
from sklearn.linear_model import LinearRegression
data['time'] = pd.to_datetime(data['time'], format='%Y-%m-%d %H:%M:%S')
data['month'] = data['time'].dt.month
data['day'] = data['time'].dt.day
data['year'] = data['time'].dt.year
train_data = data[data['time'] < '2021']
test_data = data[(data['time'] >= '2021-01-01 00:00:00') & (data['time'] < '2022-01-01 00:00:00')]
# Train a linear regression
model = LinearRegression()
model.fit(train_data[['month', 'day']], train_data['tavg'])
# Make predictions
X_train = train_data[['month', 'day']]
y_train = train_data['tmax']
X_test = test_data[['month', 'day']]
y_test = test_data['tavg']
predictions = model.predict(X_test)
print(predictions)
```

```python
from sklearn.metrics import r2_score

# Calculate R-squared score for predictions
r2 = r2_score(y_test, predictions)

# Print the R-squared score
print(f"The R-squared score of the linear regression model is: {r2}")
```

```python
def nice_axes(ax):
    ax.set_facecolor('.8')
    ax.tick_params(labelsize=8, length=0)
    ax.grid(True, axis='x', color='white')
    ax.set_axisbelow(True)
    [spine.set_visible(False) for spine in ax.spines.values()]
    
nice_axes(ax)
fig
```

```python
fig, ax_array = plt.subplots(nrows=1, ncols=4, figsize=(7, 2.5),
                             dpi=144, tight_layout=True)
dates = ['2010-03-29 00:00:00', '2015-03-30 00:00:00', '2020-03-31 00:00:00','2023-02-12 00:00:00']
for ax, date in zip(ax_array, dates):
    s = data.loc[date].sort_values()
    ax.barh(y=s.index, width=s.values, color=colors)
    ax.set_title(date, fontsize='smaller')
    nice_axes(ax)
```

```python
fig, ax_array = plt.subplots(nrows=1, ncols=3, figsize=(7, 2.5),
                             dpi=144, tight_layout=True)
dates = ['2010-01-01 00:00:00', '2015-01-01 00:00:00', '2020-01-01 00:00:00']
for ax, date in zip(ax_array, dates):
    s = data.loc[date]
    y = data.loc[date].rank(method='first')
    ax.barh(y=y, width=s.values, color=colors, tick_label=s.index)
    ax.set_title(date, fontsize='smaller')
    nice_axes(ax)
```
